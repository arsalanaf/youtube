```python
import keys
import datetime
from time import sleep
from binance.client import Client
```

```python
client = Client(keys.APIKey, keys.SecretKey) 
```

```python
ETH= client.get_historical_klines(symbol='ETHBTC', interval= '1d', start_str= '2018.01.01', end_str= '2018.03.31')
BNB= client.get_historical_klines(symbol='BNBBTC', interval= '1d', start_str= '2018.01.01', end_str= '2018.03.31')
TRX= client.get_historical_klines(symbol='TRXBTC', interval= '1d', start_str= '2018.01.01', end_str= '2018.03.31')
VEN= client.get_historical_klines(symbol='VENBTC', interval= '1d', start_str= '2018.01.01', end_str= '2018.03.31')
EOS= client.get_historical_klines(symbol='EOSBTC', interval= '1d', start_str= '2018.01.01', end_str= '2018.03.31')
```

```python
import numpy as np
import pandas as pd
```

```python
ETH_df= pd.DataFrame(ETH, columns=['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'])
BNB_df= pd.DataFrame(BNB, columns=['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'])
TRX_df= pd.DataFrame(TRX, columns=['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'])
VEN_df= pd.DataFrame(VEN, columns=['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'])
EOS_df= pd.DataFrame(EOS, columns=['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'])
```

```python
ETH_df['Open time']= pd.to_datetime(ETH_df['Open time'], unit='ms')
BNB_df['Open time']= pd.to_datetime(BNB_df['Open time'], unit='ms')
TRX_df['Open time']= pd.to_datetime(TRX_df['Open time'], unit='ms')
VEN_df['Open time']= pd.to_datetime(VEN_df['Open time'], unit='ms')
EOS_df['Open time']= pd.to_datetime(EOS_df['Open time'], unit='ms')
```

```python
ETH_df.set_index('Open time', inplace=True)
BNB_df.set_index('Open time', inplace=True)
TRX_df.set_index('Open time', inplace=True)
VEN_df.set_index('Open time', inplace=True)
EOS_df.set_index('Open time', inplace=True)
ETH_df['Close']= ETH_df['Close'].astype(float)
BNB_df['Close']= BNB_df['Close'].astype(float)
TRX_df['Close']= TRX_df['Close'].astype(float)
VEN_df['Close']= VEN_df['Close'].astype(float)
EOS_df['Close']= EOS_df['Close'].astype(float)
```

```python
ETH_df
```

```python
coins= pd.concat([ETH_df['Close'], BNB_df['Close'], TRX_df['Close'], VEN_df['Close'], EOS_df['Close']], axis=1)
coins.columns= ['ETH', 'BNB', 'TRX', 'VEN', 'EOS']
coins.corr()
```

```python
daily_return= coins.pct_change().dropna()
q1_return= daily_return.mean()*90
q1_cov= daily_return.cov()*90
```

```python
pf_returns, pf_volatility, pf_sharpe_ratio, pf_coins_weights=([] for i in range(4))
num_portfolios= 10000
for portfolio in range(num_portfolios):
    weights= np.random.random(5)
    weights /= np.sum(weights)
    returns = np.dot(weights, q1_return)
    volatility = np.sqrt(np.dot(weights.T, np.dot(q1_cov, weights)))
    sharpe = returns / volatility
    pf_coins_weights.append(weights)
    pf_returns.append(returns)
    pf_volatility.append(volatility)
    pf_sharpe_ratio.append(sharpe)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
plt.figure(figsize=(15,7))
plt.scatter(x=pf_volatility, y=pf_returns, c= pf_sharpe_ratio, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
sns.set(style='darkgrid')
plt.title('Efficient Frontier')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.show()
```
